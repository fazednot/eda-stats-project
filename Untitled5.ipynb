```python
import os

# Creating the folder structure
folders = [
    '.vscode',
    '.github/workflows',
    'src',
    'notebooks',
    'tests',
    'scripts'
]

for folder in folders:
    os.makedirs(folder, exist_ok=True)

# Creating necessary files
files = [
    '.gitignore',
    'requirements.txt',
    'README.md',
    'notebooks/README.md',
    'scripts/README.md',
    'tests/__init__.py',
    'notebooks/__init__.py',
    'scripts/__init__.py',
    '.vscode/settings.json',
    '.github/workflows/unittests.yml'
]

for file in files:
    with open(file, 'w') as f:
        pass  # This will create empty files

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('togo-dapaong_qc.csv')
df.head()
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
print("Data types of each column:\n", df.dtypes)
print("Summary statistics:\n", df.describe())

# Convert 'Timestamp' to datetime object and exclude from correlation
df['Timestamp'] = pd.to_datetime(df['Timestamp'])  # Convert to datetime
numeric_df = df.select_dtypes(include=['number'])  # Select only numeric columns

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Solar Radiation Data')
plt.show()
import matplotlib.pyplot as plt

df[['GHI', 'DNI', 'DHI']].hist(bins=30, figsize=(12, 8))
plt.suptitle('Distribution of Solar Radiation Metrics')
plt.show()
# Relationship between Ambient Temperature and GHI
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Tamb', y='GHI', data=df)
plt.title('Relationship between Ambient Temperature and GHI')
plt.xlabel('Ambient Temperature (°C)')
plt.ylabel('Global Horizontal Irradiance (W/m²)')
plt.show()
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp', inplace=True)

df[['GHI', 'DNI', 'DHI']].resample('D').mean().plot(figsize=(14, 7))
plt.title('Daily Averages of Solar Radiation Metrics Over Time')
plt.ylabel('Irradiance (W/m²)')
plt.show()
# Example: Impact of Cleaning on GHI
df.groupby('Cleaning')['GHI'].mean().plot(kind='bar', color='skyblue')
plt.title('Impact of Cleaning Events on GHI')
plt.ylabel('Average GHI (W/m²)')
plt.show()

```
